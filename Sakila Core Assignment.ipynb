## Queries

### Query 1

What query would you run to get all the customers inside city_id = 312? <br>
Your query should return the customers' first name, last name, email, address, and city.

```python
q1 = """
SELECT first_name,  last_name, email, address_id, city
FROM customer, city
WHERE city_id=312;
"""
pd.read_sql(q1, engine)
```

### Query 2

What query would you run to get all comedy films? <br>
Note that the genre is called the category in this schema. <br>
Your query should return film title, description, release year, rating, and special features.

```python
q2 = """
SELECT title, description, release_year, rating, special_features
FROM film, category
WHERE name='Comedy';
"""
pd.read_sql(q2, engine)
```

### Query 3

What query would you run to get all the films that Johnny Lollobrigida was in? <br>
Your query should return the actor's last name, film title, and release year.

```python
q3 = """
SELECT last_name, title, release_year
FROM actor, film
WHERE last_name='Lollobrigida';
"""
pd.read_sql(q3, engine)
```

### Query 4

What query would you run to get the first and last names of all <br>
the actors in the movie titled "Bingo Talented"?

```python
q4 = """
SELECT first_name, last_name, title
FROM actor, film
WHERE title='Bingo Talented';
"""
pd.read_sql(q4, engine)
```

### Query 5

What query would you run to get the customer_id associated <br>
with all payments greater than twice the average payment amount? <br> 
(HINT: use 2* in your query to get twice the amount). <br>
Your result should include the customer id and the amount.

```python
q5 = """
SELECT customer_id, amount
FROM payment
WHERE amount > 2 * (SELECT AVG(amount) FROM payment);
"""
pd.read_sql(q5, engine)
```

### Query 6

What query would you run to list the first and last names of the <br>
5 customers who have the highest number(count) of payments? <br>
You can title the number of payments as num_payments.<br>

```python
q6 = """
SELECT c.first_name, c.last_name, COUNT(p.customer_id) AS num_payments
FROM customer c
JOIN payment p ON c.customer_id = p.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name
ORDER BY num_payments DESC
LIMIT 5;
"""
pd.read_sql(q6, engine)
```
